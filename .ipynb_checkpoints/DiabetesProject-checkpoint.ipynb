```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
train_data = pd.read_csv("C:/Users/CBIT/Machine Learning Projects/Diabetes_prediction_knn/archive (1)/diabetes_012_health_indicators_BRFSS2015.csv")
```

```python
train_data.isnull().sum()
```

```python
train_data.columns
```

```python
train_data['Age'].min()
```

```python
train_data['Diabetes_012'].min()
```

```python
y = train_data['Diabetes_012']
```

```python
X = train_data.loc[:,'HighBP':]
```

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=42)
```

```python
 X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
classifier = KNeighborsClassifier(n_neighbors = 7)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_preds = classifier.predict(X_test)
```

```python
y_preds != y_test
```

```python
test_error_rate = np.mean(y_preds != y_test)
print(round(test_error_rate*100, 2))
```

```python
accuracy_score(y_preds, y_test)*100
```

```python
train_preds = classifier.predict(X_train)
```

```python
test_error_rate = np.mean(train_preds != y_train)
print(round(test_error_rate*100, 2))
```

```python
accuracy_score(train_preds, y_train)*100
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(f'Confusion Matrix : ')
print()
print(confusion_matrix(y_test, y_preds))
print()
print(f'Classification Report: ')
print()
print(classification_report(y_test, y_preds))
```

```python
import time
```

```python
error = []
start = time.time()
for i in range(1,21):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)

    error.append(np.mean(pred_i != y_test))
end = time.time()
print(error)
print(end-start)
```
