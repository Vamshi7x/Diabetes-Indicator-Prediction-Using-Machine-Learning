```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
import warnings 
warnings.filterwarnings('ignore')
import sklearn
```

```python
print('numpy', np.__version__)
print('pandas', pd.__version__)
print('sklearn', sklearn.__version__)
```

```python
df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')
```

```python
df.head()
```

```python
df = df.astype(int)
```

```python
df['Diabetes_012'].value_counts()
```

Since, the class 1 prediabetic & 2 Diabetec are same converting the class 1 & 2 to Diabetec as class 1

```python
df['Diabetes_012'][df['Diabetes_012'] == 2] = 1
df['Diabetes_012'].value_counts()
```

```python
df['Diabetes_012'].value_counts()
```

## As this is a diabetes indicator. I am considering the recall metric is best for this project

# Model Building

```python
y = df['Diabetes_012']
X = df.drop('Diabetes_012', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 73)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Logistic Regression

```python
logit = LogisticRegression()
```

```python
logit.fit(X_train,y_train)
```

```python
logit_test = logit.predict(X_test)
logit_train = logit.predict(X_train)
```

```python
print(accuracy_score(y_test,logit_test))
print(accuracy_score(y_train,logit_train))
print(confusion_matrix(y_test,logit_test))
print(classification_report(y_test,logit_test))
```

## Naive Bayes

```python
NB = GaussianNB(var_smoothing = 0.001)
```

```python
NB.fit(X_train, y_train)
```

```python
NB_test = NB.predict(X_test)
NB_train = NB.predict(X_train)
```

```python
print(accuracy_score(y_test,NB_test))
print(accuracy_score(y_train,NB_train))
print(confusion_matrix(y_test,NB_test))
print(classification_report(y_test,NB_test))
```

## KNN 

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn_test = knn.predict(X_test)
knn_train = knn.predict(X_train)
```

```python
print(accuracy_score(y_test,knn_test))
print(accuracy_score(y_train,knn_train))
print(confusion_matrix(y_test,knn_test))
print(classification_report(y_test,knn_test))
```

## Decision Tree

```python
Tree = DecisionTreeClassifier(class_weight = 'balanced')
```

```python
Tree.fit(X_train, y_train)
```

```python
tree_test = Tree.predict(X_test)
tree_train = Tree.predict(X_train)
```

```python
print(accuracy_score(y_test,tree_test))
print(accuracy_score(y_train,tree_train))
print(confusion_matrix(y_test,tree_test))
print(classification_report(y_test,tree_test))
```

## Random Forest

```python
Forest = RandomForestClassifier(n_estimators = 99, random_state = 73, n_jobs = -1, class_weight = 'balanced', oob_score = True, )
```

```python
Forest.fit(X_train, y_train)
```

```python
test_forest = Forest.predict(X_test)
train_forest = Forest.predict(X_train)
```

```python
print(accuracy_score(y_test,test_forest))
print(accuracy_score(y_train,train_forest))
print(confusion_matrix(y_test,test_forest))
print(classification_report(y_test,test_forest))
```
