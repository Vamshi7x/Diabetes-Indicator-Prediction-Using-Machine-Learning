```python
y = df['Diabetes_012']
X = df.drop('Diabetes_012', axis = 1)
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
```

```python
rus = RandomUnderSampler(random_state = 42)
```

```python
rus_X, rus_y = rus.fit_resample(X_train, y_train)
```

```python
ros = RandomOverSampler(random_state = 42)
```

```python
ros_X, ros_y = ros.fit_resample(X_train, y_train)
```

```python
smote = SMOTE(random_state = 42)
```

```python
smote_X, smote_y = smote.fit_resample(X_train, y_train)
```

# Random Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import make_scorer, recall_score, precision_score, f1_score
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler
```

```python
pipeline = Pipeline([
    ('oversample', RandomOverSampler(random_state=42)),
    ('rf', RandomForestClassifier(
        random_state=42,
        n_jobs=-1
    ))
])
```

```python
param_dist = {
    'rf__n_estimators': [200, 300, 400, 500],
    'rf__max_depth': [6, 8, 10, None],
    'rf__min_samples_split': [20, 40, 60],
    'rf__min_samples_leaf': [10, 20, 30],
    'rf__max_features': ['sqrt', 'log2']
}
```

```python
scoring = {
    'recall': make_scorer(recall_score, pos_label=1),
    'precision': make_scorer(precision_score, pos_label=1),
    'f1_macro': make_scorer(f1_score, average='macro')
}
```

```python
random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=30,
    scoring=scoring,
    refit='recall',  
    cv=5,
    verbose=2,
    random_state=42,
    n_jobs=-1
)
random_search.fit(ros_X, ros_y)
```

```python
best_model = random_search.best_estimator_

y_pred = best_model.predict(X_test)

print("Best Params:\n", random_search.best_params_)
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
print("Best CV Recall:", random_search.cv_results_['mean_test_recall'][random_search.best_index_])
print("Best CV Precision:", random_search.cv_results_['mean_test_precision'][random_search.best_index_])
print("Best CV Macro F1:", random_search.cv_results_['mean_test_f1_macro'][random_search.best_index_])
```
